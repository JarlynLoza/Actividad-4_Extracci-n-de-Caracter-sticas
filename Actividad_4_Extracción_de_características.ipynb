```python
#Importar librerías pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from  funpymodeling.exploratory import freq_tbl
```

```python
#Carga desde un archivo .xlsx sin indice
Micro_Retailer= pd.read_excel('microretailer_mit_lift_lab.xlsx')
```

```python
Micro_Retailer.head()
```

```python
#Seleccionar columnas numéricas
numericas = Micro_Retailer.select_dtypes(include=['number'])
print("Columnas numericas:", numericas)

print("************************************************************")
#Seleccionar columnas categóricas
categoricas = Micro_Retailer.select_dtypes(exclude=['number'])
print(("Columnas categoricas:", categoricas))
```

## Limpieza de valores nulos del dataframe

```python
#Reemplazamos valores nulos del dataframe con bfill
df_nnulos = categoricas.fillna(method="bfill")
data = categoricas.fillna(method="ffill")

```

```python
#Corroboramos valores nulos
df=data.isnull().sum()
df
```

```python
freq_tbl(data)
```

##Extraccion de características

**Columna 1:'232_type_of_store'**


```python
#Obtener un análisis univariado de una variable categórica en específico
T1=freq_tbl(data['232_type_of_store'])
T1
```

```python
#Obtener un filtro de los valores más reelevantes de las variables categórica seleccionada
Filtro1=T1[T1['frequency']>1]
Filtro1
```

```python
#Ajustar el índice del dataframe
Filtro1_index=Filtro1.set_index('232_type_of_store')
Filtro1_index
```

```python
#Grafico de pastel del dataframe filtrado
Filtro1_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f %%")
```
